```python
from os import listdir
from os.path import isfile, join

mypath = "/mnt/archive2/source"
subjects = listdir(mypath)
subjects.sort()

for subject in subjects:
  print("\nSubject: {}, # Days: {}".format(subject, len(listdir(join(mypath, subject)))))
  for date in listdir(join(mypath, subject)):
    print("Date: {}, # Samples: {}".format(date, len(listdir(join(mypath, subject, date)))))
```
